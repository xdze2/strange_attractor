# Where is the fractal?
Lorenz attractor in Julia

```julia
using DifferentialEquations
```

```julia
# Parameters
sigma = 10.0
rho = 28.0
beta = 8/3.;

# Attractor size
xy_half_size = 25
z_size = 45
```

```julia
# Fixed points
p1 = [-sqrt(beta*(rho-1)), -sqrt(beta*(rho-1)), rho-1];
p2 = [+sqrt(beta*(rho-1)), +sqrt(beta*(rho-1)), rho-1];
p12 = p2 .- p1
```

```julia
function lorenz!(du, u, p, t)
    du[1] = sigma*(u[2]-u[1])
    du[2] = u[1]*(rho-u[3]) - u[2]
    du[3] = u[1]*u[2] - beta*u[3]
    du = du ./ sqrt(sum(du.^2))  # norm
end

# Burn iterations to reach the attractor
function burn()
    u0 = (rand(Float64, 3) .- [0.5, 0.5, 0]).*[2*xy_half_size, 2*xy_half_size, z_size]
    tspan = (0.0, 100.0)
    prob = ODEProblem(lorenz!, u0, tspan)
    tol = 1e-3
    sol = solve(prob, Tsit5(), reltol=tol, abstol=tol,
                save_everystep=false);
    return sol[end]
end
```

```julia
# Define one orbit
u0 = burn()

# integrate for the given time, starting from saved u0
function orbit(t, u0=[-14.19305, -5.5450, 41.7958545])
    tspan = (0.0, t)
    prob = ODEProblem(lorenz!, u0, tspan)

    tol = 1e-6
    sol = solve(prob, Tsit5(),
                reltol=tol, abstol=tol, save_everystep=false);
    return sol.u[end]
end
```

```julia
using LinearAlgebra
using Plots
```

```julia
# Define a plane
function get_plane_base(p)
    n = lorenz!(zeros(3), p, 0, 0) # normal
    r = normalize(p) # another arbitrary direction 
    t_y = normalize( cross(r, n) )
    t_x = cross(t_y, n)
    return p, t_x, t_y, n
end


function projection(base, u)
    p, t_x, t_y, n = base
    a = u .- p
    ux = sum(a .* t_x)
    uy = sum(a .* t_y)
    return [ux, uy]
end

function unproj(base, uv)
    p, t_x, t_y, n = base
    return p .+  t_x.*uv[1] .+  t_y.*uv[2]
end
```

```julia
# Integrate till a given plane
function plane_callback(base)
    p, t_x, t_y, n = base
    
    negative_side(u, t, integrator) = -sum((u .- p) .* n)
    
    cb = ContinuousCallback(negative_side, nothing, affect!,
                        rootfind=true, save_positions=(false, true))
    return cb
end

function affect!(integrator)
    terminate!(integrator)
end

function random_point_on_plane(plane_cb)
    u0 = burn()
   
    tspan = (0.0, 400.0)
    prob = ODEProblem(lorenz!, u0, tspan)

    tol = 1e-5
    sol = solve(prob, Tsit5(), callback=plane_cb,
                reltol=tol, abstol=tol, save_everystep=false);
    return sol.u[end]
end
```

```julia
# Run the solver and fill the cube
function populate!(cube)

    u0 = burn()

    tspan = (0.0, 200.0)
    prob = ODEProblem(lorenz!, u0, tspan)

    tol = 1e-6
    sol = solve(prob, Tsit5(), #callback=cb,
                reltol=tol, abstol=tol, saveat=dt);

    #saved_event_idx
    #println("sol size: ", size(sol.u))

    for xyz in sol.u
        ijk = to_ijk(xyz)
        cube[ijk] = get(cube, ijk, (0, -1.0)) .+ (1, 0)
    end
    
end
```

```julia
# Init the cube
cube = Dict{Tuple{Int, Int, Int}, Tuple{Int64, Float64}}()
```

```julia
for k in 1:500
    populate!(cube)
    print("\rcube size:  ", length(cube), "\t k: ", k)
end
println("\n min, max= ", extrema([ x[1] for x in values(cube)]))
println("c min, max= ", extrema([ x[2] for x in values(cube)]))
```

```julia
# Same for the 3 coordinates
function to_xyz(ijk)
    i, j, k = ijk
    x = i*voxel_size - xy_half_size
    y = j*voxel_size - xy_half_size
    z = k*voxel_size
   return  x, y, z
end
```
